```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

from collections import Counter
```

```python
# Reading data from csv file
plt.style.use('fivethirtyeight') # Provided a plotting style

data = pd.read_csv('data.csv') # Reading csv file using pandas

# Serise -> DS Conversion
ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']
# print(type(lang_responses))
# print(lang_responses)

language_counter = Counter() # Uses HashMap algorithm -> Takes list as an argument and count them

for response in lang_responses:
    language_counter.update(response.split(';')) # update method -> updates the count of classes

# print(language_counter) # Returns a dictonary of class as key and counted_value as pair 

languages = []
popularity = []

top_languages = language_counter.most_common(15) # top counted classes -> Convert them into list of tuples

for item in top_languages:
    languages.append(item[0])
    popularity.append(item[1])

plt.barh(languages[::-1], popularity[::-1]) # Reversing them for showing top to bottom
plt.title('Top Programming Languages used by people all over the world')
plt.xlabel('No. People using PL')

plt.show()
```
